```python
df['bedrooms']
```

```python
df[['bedrooms','price','sqft_living']]
```

```python
# boolean indexing
df[df['price']>10000]
```

```python
df[df['condition']>=5]
```

```python
# 5 star rated house + 2 floors
df[(df['condition']==5)& (df['floors']==2)]
```

```python
f1=df['price']>10000
f2=df['condition']==5
f3=df['floors']==2
df[f1&f2&f3]
```

```python
df[f1&f2&f3].reset_index().style.highlight_max(color='blue')
```

```python
df[f1&f2&f3].reset_index().style.background_gradient(cmap='hot')
```

```python
# query like sql
df.query('price>10000' and 'condition==2')
```
