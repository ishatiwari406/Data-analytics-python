```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,5))
plt.plot([1,2,3],[4,5,6],linewidth=5,color='red',marker='o',markersize=20)
plt.plot([1,2,3],[6,5,4],linewidth=5,color='blue',marker='o',markersize=20)
plt.xticks([1,2,3],['one','two','three'],rotation=45)
plt.yticks([4,5,6],['four','five','six'],rotation=45)
plt.title('My first plot')
plt.xlabel('x- axis')
plt.ylabel('y-axis')
plt.show()
```
