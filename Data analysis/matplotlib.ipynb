```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(5,5))
plt.plot([1,2,3],[4,5,6],linewidth=5,color='red',marker='o',markersize=20)
plt.plot([1,2,3],[6,5,4],linewidth=5,color='blue',marker='o',markersize=20)
plt.xticks([1,2,3],['one','two','three'],rotation=45)
plt.yticks([4,5,6],['four','five','six'],rotation=45)
plt.title('My first plot')
plt.xlabel('x- axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.subplots(nrows=1,ncols=2,figsize=(10,5))
x= np.arange(-10,10,.5)
print(x)
```

```python
fig, ax=plt.subplots(nrows=1,ncols=2,figsize=(10,5)) # 10 is the width and 5 is the height
t= np.arange(-10,10,.1)
sine=np.sin(t)
cosine=np.cos(t)
#print(t)
#print(sine)
#print(cosine)
print(ax[0].plot(t,sine,label='Sine',color='red')) # ax[0] --> denotes the first graph
print(ax[1].plot(t,cosine,label='Cosine',color='g')) # ax[0] --> denotes the first graph
plt.show()
plt.savefig('Sine_cos_graph.png')
```

Graphs
- plt.plot -->line graph 
- plt.bar -->bar graph
