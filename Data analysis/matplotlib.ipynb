```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(5,5))
plt.plot([1,2,3],[4,5,6],linewidth=5,color='red',marker='o',markersize=20)
plt.plot([1,2,3],[6,5,4],linewidth=5,color='blue',marker='o',markersize=20)
plt.xticks([1,2,3],['one','two','three'],rotation=45)
plt.yticks([4,5,6],['four','five','six'],rotation=45)
plt.title('My first plot')
plt.xlabel('x- axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.subplots(nrows=1,ncols=2,figsize=(10,5))
x= np.arange(-10,10,.5)
print(x)
```

```python
fig, ax=plt.subplots(nrows=1,ncols=2,figsize=(10,5)) # 10 is the width and 5 is the height
t= np.arange(-10,10,.1)
sine=np.sin(t)
cosine=np.cos(t)
#print(t)
#print(sine)
#print(cosine)
print(ax[0].plot(t,sine,label='Sine',color='red')) # ax[0] --> denotes the first graph
print(ax[1].plot(t,cosine,label='Cosine',color='g')) # ax[0] --> denotes the first graph
plt.show()
plt.savefig('Sine_cos_graph.png')
```

Graphs
- plt.plot -->line graph 
- plt.bar -->bar graph
- plt.pie -->pie chart
- plt.scatter -->scatter graph
- plt.hist -->histogram

```python
fig,((g1,g2),(g3,g4))=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
days=['sun','mon','tue','wed','thu','fri','sat']
sales=[160,155,200,450,120,360,400]
profit=np.random.normal(50,20,100) #mean=50, std. deviation=20, samples=100
g1.bar(days,sales)
g2.hist(profit, bins=[0,20,40,60,80,100],color='g',edgecolor='black')
g3.pie(sales,labels=days,autopct='%.1f%%',wedgeprops={'width':0.7})
x=np.random.normal(0,1,500)
y=np.random.normal(0,1,500)
g4.scatter(x,y,color='orange',alpha=0.5)
plt.show()
```
